```python
#Filling missing values - Mean for Salary
df['Sal'].fillna(df['Sal'].mean(), inplace = True)
```

```python
#Filling missing values - Median for Experience
df['Experience'].fillna(df['Experience'].median(), inplace = True)
```

```python
#Filling missing values - Mode for performance Score
df['Per_score'].fillna(df['Per_score'].mode()[0], inplace = True)
```

```python
#Filling missing values - Median for Age
df['Age'].fillna(df['Age'].median(), inplace = True)
```

```python
#Filling missing values - Mode for department
df['Dept'].fillna(df['Dept'].mode()[0],inplace = True)
```

```python
# Convert Joining_Date to datetime
df['Joining_Date'] = pd.to_datetime(df['Joining_Date'],errors = 'coerce')
```

```python
#Remove duplicates
df.drop_duplicates(inplace = True)
```
