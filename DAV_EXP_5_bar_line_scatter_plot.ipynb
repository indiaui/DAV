```python
# 1. Histogram: Age Distribution of Students
plt.figure(figsize=(8, 6))
sns.histplot(df['age'], kde=True, bins=8, color='skyblue')
plt.title('Age Distribution of Students')
plt.xlabel('Age')
plt.ylabel('Frequency')
```

```python
# 2. Bar Plot: Average Final Grade by Family Size
plt.figure(figsize=(8, 6))
sns.barplot(x='famsize', y='G3', data=df, palette='pastel')
plt.title('Average Final Grade by Family Size')
plt.xlabel('Family Size (LE3: <=3, GT3: >3)')
plt.ylabel('Average Final Grade (G3)')
```

```python
# 3. Line Plot: Trend of Absences by Age
plt.figure(figsize=(10, 6))
sns.lineplot(x='age', y='absences', data=df, marker='o', ci=None)
plt.title('Average Absences by Student Age')
plt.xlabel('Age')
plt.ylabel('Average Number of Absences')
plt.grid(True)
```

```python
# 4. Scatter Plot: First Period Grade vs. Final Grade
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='G1', y='G3', hue='internet', style='internet', s=80)
plt.title('First Period Grade (G1) vs. Final Grade (G3)')
plt.xlabel('First Period Grade (G1)')
plt.ylabel('Final Grade (G3)')
```
