```python
# Check unique entries in key columns
for col in ["State", "Product", "Sales Channel"]:
    print(f"{col}: {df[col].unique()}")
```

```python
#Filling missing value of 'Units Sold' with 0
df['Units Sold'] = df['Units Sold'].fillna(0)

#Filling missing values of 'Unit Price' with Median
df['Unit Price'] = df['Unit Price'].fillna(df['Unit Price'].median())

#Filling missing value of 'Order ID' with 0
df['Order ID'] = df['Order ID'].fillna(0)
```

```python
df['Total Sale'] = df['Units Sold']*df['Unit Price']
```

```python
#Filling missing value of 'Sales Channel' with its mode
df['Sales Channel'] = df['Sales Channel'].fillna(df['Sales Channel'].mode()[0])
```

```python
#Check for missing values
df.isnull().sum()
```

```python
# Example: standardize state names
df["State"] = df["State"].str.strip().str.title()

# Example: normalize product categories
df["Product_Category"] = df["Product"].str.strip().str.title()

# Example: normalize sales channels
df["Sales_Channel"] = df["Sales Channel"].str.strip().str.title()
```

```python
df
```

```python
# Convert to datetime
df["Sale Date"] = pd.to_datetime(df["Sale Date"], errors="coerce")
```

```python
# Fill missing dates if needed
df['Sale Date'] = df["Sale Date"].ffill()
```

```python
# Extract useful date parts
df["Year"] = df["Sale Date"].dt.year
df["Month"] = df["Sale Date"].dt.month_name()
df["Day"] = df["Sale Date"].dt.day
```
