```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
#Load your dataset
df = pd.read_csv('/content/Mall_Customers_Enhanced.csv')
```

```python
  #Print Dataset
  df
```

```python
# calculate statistics
mean = df.mean(numeric_only=True)
median = df.median(numeric_only=True)
mode = df.mode(numeric_only=True).iloc[0]
std = df.std(numeric_only=True)
```

```python
print("Mean : ",mean)
print("Median : ",median)
print("Mode : ",mode)
print("Standard Deviation : ",std)
```

```python
# Label encoding for 'Preferred Category'
if 'Preferred Category' in df.columns:
  le = LabelEncoder()
  df['Preferred Category_encoded'] = le.fit_transform(df['Preferred Category'])
  print(df[['Preferred Category', 'Preferred Category_encoded']].head(50))
```

```python
#One Hot Encoding for 'Preferred Category'
one_hot_encoded = pd.get_dummies(df['Preferred Category'], prefix='Preferred Category')
df = pd.concat([df, one_hot_encoded], axis=1)
df
```

```python
#One Hot Encoding for 'Preferred Category'
pd.set_option('display.width', 1000) #Increase Width
pd.set_option('display.max_columns', None) #Show all columns
pd.set_option('display.max_rows', None) #Show all rows
if 'Preferred Category' in df.columns:
  category_ohe = pd.get_dummies(df['Preferred Category'], prefix='Preferred Category')
  print(category_ohe.head(10))
```

```python
statistics = df[['Annual Income (k$)','Spending Score (1-100)', 'Estimated Savings (k$)','Credit Score']].agg([ 'mean', 'median'])
mode_values = df[['Annual Income (k$)','Spending Score (1-100)', 'Estimated Savings (k$)','Credit Score']].mode().iloc[0] # mode may have multiple, take first
# Combine results
statistics.loc['mode'] = mode_values
# Show as single row (flatten)
result = statistics.transpose().reset_index()
result.columns = ['Column','Mean','Median','Mode']
print(result)
```
