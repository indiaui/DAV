```python
# 1. Heatmap: Correlation of Numerical Features
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
plt.figure(figsize=(18, 14))
correlation_matrix = df[numeric_cols].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Student Data')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
```

```python
# 2. Violin Plot: Distribution of Final Grades by Study Time
plt.figure(figsize=(10, 7))
sns.violinplot(x='studytime', y='G3', data=df, palette='muted')
plt.title('Distribution of Final Grades by Weekly Study Time')
plt.xlabel('Weekly Study Time (1: <2, 2: 2-5, 3: 5-10, 4: >10 hrs)')
plt.ylabel('Final Grade (G3)')
```
